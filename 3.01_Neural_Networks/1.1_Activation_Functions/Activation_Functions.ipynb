```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn-v0_8-darkgrid")
x = np.linspace(-10, 10, 200)
```

#### **1. Sigmoid Activation Function**


**Sigmoid:**

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

- **Sigmoid Output Range:** (0, 1)


```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
sig = sigmoid(x)

plt.figure(figsize=(8, 5))
plt.plot(x, sig, label="Sigmoid")
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("σ(x)")
plt.legend()
plt.show()
```

**Sigmoid Derivative:**

$$
\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))
$$

- **Derivative Output Range:** (0, 0.25)


```python
def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)
```

```python
sig_deriv = sigmoid_derivative(x)

plt.figure(figsize=(8, 5))
plt.plot(x, sig_deriv, label="Sigmoid Derivative", color="red")
plt.title("Sigmoid Derivative")
plt.xlabel("x")
plt.ylabel("σ'(x)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, sig, label="Sigmoid")
plt.plot(x, sig_deriv, label="Sigmoid Derivative", color="red")
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("σ(x) , σ'(x)")
plt.legend()
plt.show()
```

#### **2. Tanh (Hyperbolic Tangent) Activation Function**


**Tanh:**

$$
\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

- **Tanh Output Range:** (-1, 1)


```python
def tanh(x):
    return np.tanh(x)
```

```python
tanh_ = tanh(x)

plt.figure(figsize=(8, 5))
plt.plot(x, tanh_, label="Tanh")
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.legend()
plt.show()
```

**Tanh Derivative:**

$$
\text{tanh}'(x) = 1 - \text{tanh}^2(x)
$$

- **Tanh Derivative Output Range:** (0, 1)


```python
def tanh_derivative(x):
    return 1 - (np.tanh(x) ** 2)
```

```python
tanh_deriv = tanh_derivative(x)

plt.figure(figsize=(8, 5))
plt.plot(x, tanh_deriv, label="Tanh Derivative", color="red")
plt.title("Tanh Derivative")
plt.xlabel("x")
plt.ylabel("tanh'(x)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, tanh_, label="Tanh")
plt.plot(x, tanh_deriv, label="Tanh Derivative", color="red")

plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x) , tanh'(x)")

plt.legend()
plt.show()
```

#### **3. ReLU (Rectified Linear Unit) Activation Function**


**ReLU:**

$$
\text{ReLU}(x) = \max(0, x)
$$

- **ReLU Output Range:** [0, ∞)


```python
def relu(x):
    return np.maximum(0, x)
```

```python
relu_ = relu(x)

plt.figure(figsize=(8, 5))
plt.plot(x, relu_, label="ReLU")
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.legend()
plt.show()
```

**ReLU Derivative:**

$$
\text{ReLU}'(x) = \begin{cases}
1 & \text{if } x > 0 \\
0 & \text{if } x \leq 0
\end{cases}
$$


```python
def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
relu_deriv = relu_derivative(x)

plt.figure(figsize=(8, 5))
plt.plot(x, relu_deriv, label="ReLU Derivative", color="red")
plt.title("ReLU Derivative")
plt.xlabel("x")
plt.ylabel("ReLU'(x)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, relu_, label="ReLU")
plt.plot(x, relu_deriv, label="ReLU Derivative", color="red")

plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x) , ReLU'(x)")

plt.legend()
plt.show()
```

#### **4. Leaky ReLU Activation Function**


**Leaky ReLU:**

$$
\text{Leaky ReLU}(x) = \begin{cases}
x & \text{if } x > 0 \\
\alpha x & \text{if } x \leq 0
\end{cases}
$$

- **Leaky ReLU Output Range:** (-∞, ∞)


```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, (alpha * x))
```

```python
leaky_relu_ = leaky_relu(x)

plt.figure(figsize=(8, 5))
plt.plot(x, leaky_relu_, label="Leaky ReLU")
plt.title("Leaky ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("Leaky_ReLU(x)")

plt.legend()
plt.show()
```

**Leaky ReLU Derivative:**

$$
\text{Leaky ReLU}'(x) = \begin{cases}
1 & \text{if } x > 0 \\
\alpha & \text{if } x \leq 0
\end{cases}
$$

- **where \( α \) is a small positive number like 0.01**


```python
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)
```

```python
leaky_relu_deriv = leaky_relu_derivative(x)

plt.figure(figsize=(8, 5))
plt.plot(x, leaky_relu_deriv, label="Leaky ReLU Derivative", color="red")
plt.title("Leaky ReLU Derivative")
plt.xlabel("x")
plt.ylabel("Leaky_ReLU'(x)")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, leaky_relu_, label="Leaky ReLU")
plt.plot(x, leaky_relu_deriv, label="Leaky ReLU Derivative", color="red")
plt.title("Leaky ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("Leaky_ReLU(x) , Leaky_ReLU'(x)")

plt.legend()
plt.show()
```

#### **5. Softmax Activation Function**


**Softmax:**

$$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{K} e^{x_j}}
$$

- **Softmax Output Range:** (0, 1) for each class, with the sum of outputs equal to 1.


```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x, axis=0, keepdims=True)
```
