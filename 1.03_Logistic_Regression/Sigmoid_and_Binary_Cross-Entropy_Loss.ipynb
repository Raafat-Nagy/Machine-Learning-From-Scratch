# **Binary Classification: Sigmoid and Binary Cross-Entropy Loss**


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

---

## **Sigmoid Function (Ïƒ)**
The sigmoid function converts a real-valued logit (**z**) into a probability (**y'**) between 0 and 1:

$$
y' = \sigma(z) = \frac{1}{1 + e^{-z}}
$$





| Property | Formula | Description |
|----------|---------|-------------|
| Output Range | $ (0, 1) $ | Perfect for probabilities |
| At z=0 | $ \sigma(0) = 0.5 $ | Neutral prediction |
| Derivative | $ \sigma'(z) = \sigma(z)(1 - \sigma(z)) $ | Used in gradient descent |



### **Implementation**


```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Example
z = np.array([-2, 0, 2])
print("Sigmoid outputs:", sigmoid(z))
```


### **Visualization Sigmoid Function**


```python
z = np.linspace(-5, 5, 100)
plt.plot(z, sigmoid(z))
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("Ïƒ(z)")
plt.grid(True)
plt.show()
```

### **Binary Cross-Entropy (Log Loss)**

Binary Cross-Entropy (BCE), is the most common loss function for binary classification problems. It measures the difference between predicted probabilities and true binary labels.


$$
J(\mathbf{w}, b) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)}) \right]
$$

---

- **When $ y = 1 $:**  
  Loss = $ -\log(\hat{y}) $.  
  - If $ \hat{y} \approx 1 $ (correct prediction), loss $ \approx 0 $.  
  - If $ \hat{y} \approx 0 $ (wrong prediction), loss $ \rightarrow \infty $.  

- **When $ y = 0 $:**  
  Loss = $ -\log(1 - \hat{y}) $.  
  - If $ \hat{y} \approx 0 $ (correct prediction), loss $ \approx 0 $.  
  - If $ \hat{y} \approx 1 $ (wrong prediction), loss $ \rightarrow \infty $.  


---



### **Visualization Binary Cross-Entropy Loss**


```python
# Define a range of predicted probabilities from 0.01 to 0.99
y_pred_range = np.linspace(0.01, 0.99, 100)
```

```python
# Compute loss for y = 0 and y = 1
loss_y0 = -np.log(1 - y_pred_range)  # When actual label y = 0
loss_y1 = -np.log(y_pred_range)  # When actual label y = 1
```

```python
plt.figure(figsize=(10, 5))
plt.suptitle("Binary Cross-Entropy Loss Function", fontsize=14)

# Plot for Loss when y = 0
plt.subplot(1, 2, 1)
plt.plot(y_pred_range, loss_y0, label="Loss when y = 0", color="red")
plt.ylabel("Binary Cross-Entropy Loss")
plt.xlabel("Predicted Probability (Å·)")
plt.title("Loss for y = 0")
plt.grid(True)
plt.legend()

# Plot for Loss when y = 1
plt.subplot(1, 2, 2)
plt.plot(y_pred_range, loss_y1, label="Loss when y = 1", color="green")
plt.xlabel("Predicted Probability (Å·)")
plt.title("Loss for y = 1")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
